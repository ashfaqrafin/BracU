```python
# Task-1

class Customer:
    def __init__(self):
        self.ticket = []
        self.total = 0
        print('Welcome to ABC Memorial Park')

    def buyTicket(self, name, age):
        if len(self.ticket) == 3:
            print(f"You can't buy more than 3 tickets")
            return
        if age <= 10:
            self.ticket.append(50)
            self.total += 50
        else:
            self.ticket.append(100)
            self.total += 100
        print(f"Successfully purchased a ticket for {name}!")
    def showDetails(self):
        print(f"Amount of tickets: {len(self.ticket)}")
        print(f"Total price: {self.total} Taka")



print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
# Task-2

class Panda:
    def __init__(self,name,gender,age):
        self.name = name
        self.gender=gender
        self.age=age
    def sleep(self,hours):
        # food="Bamboo leaves"
        if(3<=hours<=5):
            food="Mixed Veggies"
        elif(6<=hours<=8):
            food="Eggplant & Tofu"
        elif(9<=hours<=11):
            food="Broccoli Chicken"
        else:
            return f"{self.name}'s duration is unknown thus should have only bamboo leaves"
        
        return "{} sleeps {} hours and should have {}".format(self.name,hours,food)


panda1 = Panda("Kunfu", "Male", 5) 
panda2 = Panda("Pan Pan", "Female",3) 
panda3 = Panda("Ming Ming", "Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================") 
print(panda2.sleep(10)) 
print(panda1.sleep(4)) 
print(panda3.sleep(13))
```

```python
# Task-3
class GreenPhone:
    def __init__(self, model_name, android_version, num_cameras):
        self.company = 'GreenPhone'
        self.model_name = model_name
        self.android_version = android_version
        self.num_cameras = num_cameras
        self.series = model_name[0]
        self.updates_remaining = {'A': 2, 'M': 3, 'U': 4}[self.series]

    def showSpecification(self):
        print(f"Phone Company: {self.company}")
        print(f"Model Name: {self.model_name}")
        print(f"Android Version: {self.android_version}")
        print(f"Number of Cameras: {self.num_cameras}")

    def updatePhone(self):
        if self.updates_remaining > 0:
            self.android_version += 1
            self.updates_remaining -= 1
            print(f"Your phone {self.company} {self.model_name} is upgraded to Android Version: {self.android_version}.")
        else:
            print(f"Your phone {self.company} {self.model_name} is already up to date.")



print('1=======================')
p1 = GreenPhone('A1', 12, 3)
p2 = GreenPhone('M11', 12, 4)
p3 = GreenPhone('U20', 12, 5)
p1.showSpecification()
print('2=======================')
p2.showSpecification()
print('3=======================')
p1.updatePhone()
print('4=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('5=======================')
p1.updatePhone()
p2.updatePhone()
p3.updatePhone()
print('6=======================')
p2.updatePhone()
p3.updatePhone()
print('7=======================')
p1.showSpecification()
p3.showSpecification()
```

```python
# Task-4


class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}
    def calculateGPA(self, courses, semester):
        self.grades[semester] = {}
        temp=[]
        sum=0
        for course in courses:
            courseName, grade = course.split(':')
            temp.append(courseName)
            sum+=float(grade)
        self.grades[semester][tuple(temp)]=round(sum/len(courses),2)
    def printDetails(self):
        print(f'Name: {self.name}')
        print(f'ID: {self.id}')
        for key,value in self.grades.items():
            print(f'Courses taken in {key}:')
            for key2,value2 in value.items():
                for name in key2:
                    print(name)
                print(f'GPA: {value2}')
    

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0',
'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'],
'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('---------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7',
'ENG101: 4.0'], 'Summer2022')
print('---------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('---------------------------------')
s2.printDetails()

```
