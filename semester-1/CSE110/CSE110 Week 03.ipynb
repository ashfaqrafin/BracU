## **while** loop


*   Value Initialization
*   Condition
*   Block of Code
*   Increment or Decrement

```python
count = 0 # Value Initialization
while count < 5: # Condition
    # Block of Code
    print(count,"Printing inside while block")
    count += 1 # Increment
print("Code Ended!") # code line after while block
```

```python
count = 4 # Value Initialization
while count > -1: # Condition
    # Block of Code
    print(count,"Printing inside while block")
    count -= 1 # Decrement
print("Code Ended!") # code line after while block
```

```python
# condition in loop
count = 0
while count < 10:
    if count % 2 == 0:
        print(count)
    else:
        pass
    count += 1
```

```python
n = int(input("Enter number of loops: "))

count = 0

sum_even = 0
sum_odd = 0
count_even = 0
count_odd = 0
# while count < 5:
while count < n:
  user_input = int(input("Enter a number: "))
  if user_input % 2 == 0:
      sum_even += user_input
      count_even += 1
  else:
      sum_odd +=  user_input
      count_odd += 1
  count += 1
print("sum_even:",sum_even)
print("Average(Even):", sum_even/count_even)
print("Sum of odd:",sum_odd)
print("Average(Odd):", sum_odd/count_odd)




```

### Use of **pass**

```python
count = 0
while count < 10:
    if count % 2 == 0:
        pass

    count += 3
print("Code Ended!")
print(count)
```

### Use of **break**

```python
count = 0
while count < 10:
    if count == 5:
        break
        print("Printing after break")

    else:
        print(count,"Printing inside while loop")
    count += 1
print("Code Ended!")
print(count)
```

### Use of **continue**

```python
count = 0
while count < 10:
    if count == 5:
        count += 1
        continue
        print("Print after continue")
    else:
        print(count,"printing in while -> else")
    count += 1
print("Code Ended!")
print(count)
```

### Infinity loop due to **inc/dec issue**

```python
count = 9 # Value Initialization
while count >= 0: # Condition
    # Block of Code
    print(count,"Printing inside while block")
    count += 1 # Decrement
print("Code Ended!") # code line after while block
```

### Infinity loop using **True**

```python
total = 0
while True:
    num = int(input())
    if num < 0:
        break
    else:
        total += num
print(total)
```

### Nested Loop

image.png

```python
outer = 1
while outer <= 3:
  inner = 0
  while inner < 3:
      print(outer, inner, sep = "=>")
      inner+=1
  print(f"Inner loop terminates {outer} time(s)")
  outer+=1
print("Outer loop terminates")
```

```python
# Star Pattern Problem 1 (Rectangle)
# Using while loop

total_row = 4 # int(input())
total_col = 6 # int(input())

current_row = 1
while current_row <= total_row:
    current_col = 1
    while current_col <= total_col:
        print(current_col, end="")
        current_col += 1
    print()
    current_row += 1
```

## For Loop
range function: <br>
*range(start, stop, step)* <br>
returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number. <br>

range(0,5,1) => range(0,5,1) =>  0  1  2  3  4
range(0,5)   => range(0,5,1) =>  0  1  2  3  4
range(5)     => range(0,5,1) =>  0  1  2  3  4

```python
for count in range(0,10,1):
  print(count, "Printing inside for block")
print("End of code")
```
