Write a Python function that takes two strings (case insensitive) as parameters. Your task is to concat the two strings except the common characters present in the strings. Print the modified string (lowercase) inside the function.

Function Call 1: modified_concat("aBcd", "bdGc")

Sample Output1: ag

Function Call 2: modified_concat("aqrde", "pAgrc")

Sample Output 2: qdepgc

Function Call 3: modified_concat("TonY", "tHor")

Sample Output 3: nyhr

````python
def modified_concat(st1, st2):
  str1 = st1.lower()
  str2 = st2.lower()
  result = ""

  for i in range(len(str1)):
    if str1[i] not in str2:
      result += str1[i]
  for i in range(len(str2)):
    if str2[i] not in str1:
      result += str2[i]
  print(result)

modified_concat("aBcd", "bdGc")
````

Permutation

n! / (n-r)!

````python
def factorial(num):
  result = 1
  if num != 0:
    for i in range(1, num+1):
      result *= i
  return result

def permutation(n, r):
  numerator = factorial(n)
  # for i in range(1,n+1):
  #   numerator *= i
  # print(numerator)
  # denominator = 1
  # for i in range(1, n-r+1):
  #   denominator *= i
  # print(denominator)
  denominator = factorial(n-r)
  print(numerator //denominator)

permutation(5,3)
````

````python
type_of = "Siamese"

def cat(name):
  #Inside the scope of function
  # type_of = "Persian"
  print("My cat is named", name)
  print("It is a", type_of)


#Outside the scope of function
cat("Theo")
print(type_of, "outside function")

# print(name)
````

````python
type_of = "Siamese"

def cat(name):
  #Inside the scope of function
  # type_of = "Persian"
  type_of = type_of + " cat"
  print("My cat is named", name)
  print("It is a", type_of)


#Outside the scope of function
cat("Theo")
print(type_of)
````

````python
type_of = "Siamese"

def cat(name):
  #Inside the scope of function
  global type_of
  type_of  = type_of + " cats"
  print("My cat is named", name)
  print("It is a ", type_of)


#Outside the scope of function
cat("Theo")
print(type_of)
````

````python
a = "global a"
def outer_f():
    # a = "out a"
    def inner_f():
        # a = "in a"
        print(a,"inner_f")
    inner_f()
    print(a,"outer_f")
outer_f()
print(a)
````

````python
a = "global a"
def outer_f():
  a = "outer a"
  def inner_f():
    # nonlocal a
    global a
    a = a + " in "
    print(a)
  inner_f()
  print(a)
outer_f()
print(a)
````

## Recursion

A function that calls itself


*   Has a stopping point/Base case
*   Recursive call



````python
def printNumbers(N):
  counter = N
  while counter >= 0:
    print(counter)
    counter = counter - 1

printNumbers(5)
````

````python
def printNumbersV2(N):
  print(N)
  printNumbersV2(N-1)

printNumbersV2(5)
````

````python
def printNumbersV2(N):
  print(N)
  if N > 0:
    printNumbersV2(N-1)

printNumbersV2(5)
````

````python
# def printNumbers(start, stop, step):     #iterative apporach
#   counter = start
#   while counter < stop:
#     print(counter)
#     counter = counter + step
# printNumbers(0, 5, 1)

def printNumbersRec(start, stop, step):   #recursive approach

  if start < stop:              # Stopping point or base case
    # print(start)
    # start += 1
    printNumbersRec(start+1, stop, step)
    print(start)

printNumbersRec(0, 5, 1)
````

````python
# def printReverse(start, stop, step):     #iterative apporach
#   counter = stop
#   while counter > start:
#     print(counter)
#     counter = counter - step
# printReverse(0, 5, 1)

def printReverseRec(start, stop, step):   #recursive approach

  if stop > start:              # Stopping point or base case
    print(stop)
    printReverseRec(start, stop-1, step)
    # print(stop)

printReverseRec(0, 5, 1)
````

When writing a recursive implementation of any computational problem you have to focus on a few things.

*   First, find out the mathematical formula for induction of the same function
*   Second, determine when you are supposed to terminate the recursion and what value should you return at that point.


````python
def recursiveAdd(num):
    if num == 1:
        return 1
    else:
        # temp = recursiveAdd(num-1)
        # sum = temp + num
        # return sum
        return num + recursiveAdd(num-1)
print(recursiveAdd(3))
````
