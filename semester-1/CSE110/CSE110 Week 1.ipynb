# **Variable**
Variables are containers for storing data values.

Screenshot 2023-05-30 212757.png

In python, we do not need to declare data types directly. The data type is decided by the interpreter during the run-time. Python is dynamically typed.

#### **Names Can Have:**
* Can have letters (A-Z and a-z), digits(0-9), and underscores (_).
* should maintain snake_casing. That means each word should be separated by underscores(_)
* cannot begin with digits
* cannot have whitespace and special signs (e.g.: +, -, !, @, $, #, %.)
* are case sensitive meaning ABC, Abc, abc are three different variables.
* have to be meaningful with the data stored in it.
* cannot be too lengthy and too general, need to have a balance
* should not be written with single alphabets. Minimum length should be 3.
* cannot be a reserved keyword for Python. There are a total of 35 keywords.
image.png

#### **Naming Convention:**
* don't use single alphabets
* don't use capital letter to start
* Variable names should be meaningful
* Never use the characters l (lowercase letter el), O (uppercase letter oh), or I (uppercase letter eye) as single character variable names.

# Type Conversion or Type Casting
Modifying one type of data to another. It can either be **Implicit** or **Explicit**.
* any operation with float type data will yields result in float since omission of data after decimal might give the wrong output.
* explicitly convert to str() or int() or float()
* for int() conversion data has to only contain numeric digits if str->int and for float->int the value after decimal is discarded.
* other conversions like list->tuple
* important for taking inputs.


```python
# value = int("123A")
# print(type(value))
# print(value // 10)
var1 = int(23.9999)
print(var1)
var2 = int(23)
print(var2)
var3 = str(23.9999)
print(type(var3))
print(float(var3))
```
