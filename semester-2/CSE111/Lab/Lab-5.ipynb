```python
# Task-2

class Farmer:
    def __init__(self,name_id=None):
        self.crops=[]
        self.fishes=[]
        if name_id==None:
            print("Welcome to your farm!")
        elif type(name_id)==str:
            print(f"Welcome to your farm, {name_id}!")
        elif type(name_id)==int:
            print(f"Welcome to your farm. Your farm ID is {name_id}!")
    def addCrops(self,*crops):
        if(len(crops)):
            self.crops.extend(crops)
            print(f"{len(crops)} crop(s) added")
        else:
            print("No crop added")
    
    def addFishes(self,*fishes):
        if(len(fishes)):
            self.fishes.extend(fishes)
            print(f"{len(fishes)} fish(s) added")
        else:
            print("No fish added")
    def showGoods(self):
        if(len(self.crops)):
            print(f"You have {len(self.crops)} crop(s):")
            for i in range(len(self.crops)):
                if(i==len(self.crops)-1):
                   print(self.crops[i])
                else:
                   print(self.crops[i],end=", ") 
        else:
            print("You don't have any crop(s).")
        if(len(self.fishes)):
            print(f"You have {len(self.fishes)} fish(s):")
            for i in range(len(self.fishes)):
                if(i==len(self.fishes)-1):
                   print(self.fishes[i])
                else:
                   print(self.fishes[i],end=", ")
        else:
            print("You don't have any fish(s).")




f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
# Task-4

class Sphere:
    def __init__(self,id,radius=1,color="White"):
        self.id=id
        self.radius=radius
        self.color=color
        self.Volume=(4/3)*3.1416*(self.radius**3)

    def merge_sphere(self,*spheres):
        print("Spheres are being merged")
        self.color="Mixed Color"
        for sphere in spheres:
            self.Volume+=sphere.Volume

    def printDetails(self):
        print(f"Sphere ID: {self.id}")
        print(f"Color: {self.color}")
        print(f"Volume: {self.Volume}")

sphere1 = Sphere("Sphere 1") 
print("1***************") 
sphere1.printDetails() 
print("2***************")
sphere2 = Sphere("Sphere 2", 3) 
print("3***************") 
sphere2.printDetails() 
print("4***************")
sphere3 = Sphere("Sphere 3", 2) 
print("5***************") 
sphere3.printDetails() 
print("6***************") 
sphere3.merge_sphere(sphere1,sphere2) 
print("7***************") 
sphere3.printDetails() 
print("8***************")
sphere4 = Sphere("Sphere 4", 5, "Purple") 
print("9***************") 
sphere4.merge_sphere(sphere3)
print("10***************") 
sphere4.printDetails()
```

```python
# Home Task-3

class Department:
    def __init__(self,name="ChE Department",sections=5):
        self.name=name
        self.sections=sections
        self.average=0
        print(f"The {name} has {sections} sections.")
    def add_students(self,*numbers):
        if(len(numbers)==self.sections):
            total=0
            for number in numbers:
                total+=number
            self.average=round(total/self.sections,2)
            print(f"The {self.name} has an average of {self.average} students in each section.")
        else:
            print(f"The {self.name} doesn't have {len(numbers)} sections.")
    def merge_Department(self,*departments):
        for dept in departments:
            print(f"{dept.name} is merged to {self.name}.")
            self.average=round(((self.average*self.sections)+(dept.sections*dept.average))/(self.sections),2)
        return f"Now the {self.name} has an average of {self.average} students in each section."

    
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22,
32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,
51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
# Home Task-4
class Shopidify:
    def __init__(self,name="Guest"):
        if(name!="Guest"):
            print(f"Welcome {name} to Shopidify") 
        else:
            print("Welcome to Shopidify")
        self.name=name
        self.cart={}
        self.history=[]
    def add_to_cart(self,*items):
        if(type(items[0])==list):
            for i in range(0,len(items[0]),2):
                if(items[0][i] in self.cart):
                    self.cart[items[0][i]]+=items[0][i+1]
                else:
                    self.cart[items[0][i]]=items[0][i+1]
        else:
            if(len(items)==1):
                if(items[0] in self.cart):
                    self.cart[items[0]]+=1
                else:
                    self.cart[items[0]]=1
            else:
                if(items[0] in self.cart):
                    self.cart[items[0]]+=items[1]
                else:
                    self.cart[items[0]]=items[1]
    def checkout(self):
        print(f"Checkout completed for {self.name}")
        self.history.append(self.cart)
        self.cart={}
    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for key,value in self.cart.items():
            print(f"-{key}:{value}x")
    def display_history(self):
        print(f"Purchase history for {self.name}:")
        for i in range(len(self.history)):
            print(f"Transaction {i+1}:")
            for key,value in self.history[i].items():
                print(f"-{key}:{value}x")



            

guest_account = Shopidify() 
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
john_account = Shopidify("John") 
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
guest_account.add_to_cart("Air Jordan", 2) 
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart() 
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2]) 
john_account.display_cart() 
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
guest_account.add_to_cart("Air Jordan") 
guest_account.display_cart() 
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
guest_account.checkout() 
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
guest_account.display_history() 
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
john_account.checkout() 
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx") 
john_account.display_history() 
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
```

```python
class Lab4:
    def __init__(self):
        self.x = 3
        self.y = 2
        self.sum = 5
        print(f'Initialized: x={self.x}, y={self.y}, sum={self.sum}')

    def methodA(self, x):
        print(f'methodA called with x={x}')
        self.y = self.sum + self.x - x
        print(f'Updated self.y to {self.y}')
        self.sum = x - self.y
        print(f'Updated self.sum to {self.sum}')
        d = Lab4()
        d.sum = self.sum + self.methodB(d)
        print(f'Created new Lab4 instance d with sum={d.sum}')
        print(self.x, self.y, self.sum)
        return d

    def methodB(self, t, z = 4):
        print(f'methodB called with t={t} and z={z}')
        y=2
        t.x = self.x + self.sum
        print(f'Updated t.x to {t.x}')
        y = y + t.x - t.y
        print(f'Updated y to {y}')
        self.sum = t.x + t.y + y - z
        print(f'Updated self.sum to {self.sum}')
        if z == 4:
            return y
        print(t.x, t.y, self.sum)
        p = t.methodA(y)
        print(t.x, self.y, p.sum)

obj = Lab4()
print('Created obj')
obj2 = obj.methodA(4)
print('Created obj2')
obj.methodB(obj2, 10)
print('Called methodB on obj')
```
