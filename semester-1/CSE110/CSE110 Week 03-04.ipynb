## **while** loop


*   Value Initialization
*   Condition
*   Block of Code
*   Increment or Decrement

```python
count = 0 # Value Initialization
while count < 5: # Condition
    # Block of Code
    print(count,"Printing inside while block")
    count += 1 # Increment
print("Code Ended!") # code line after while block
```

```python
count = 4 # Value Initialization
while count > -1: # Condition
    # Block of Code
    print(count,"Printing inside while block")
    count -= 1 # Decrement
print("Code Ended!") # code line after while block
```

```python
# condition in loop
count = 0
while count < 10:
    if count % 2 == 0:
        print(count)
    else:
        pass
    count += 1
```

```python
n = int(input("Enter number of loops: "))

count = 0

sum_even = 0
sum_odd = 0
count_even = 0
count_odd = 0
# while count < 5:
while count < n:
  user_input = int(input("Enter a number: "))
  if user_input % 2 == 0:
      sum_even += user_input
      count_even += 1
  else:
      sum_odd +=  user_input
      count_odd += 1
  count += 1
print("sum_even:",sum_even)
print("Average(Even):", sum_even/count_even)
print("Sum of odd:",sum_odd)
print("Average(Odd):", sum_odd/count_odd)




```

### Use of **pass**

```python
count = 0
while count < 10:
    if count % 2 == 0:
        pass

    count += 3
print("Code Ended!")
print(count)
```

### Use of **break**

```python
count = 0
while count < 10:
    if count == 5:
        break
        print("Printing after break")

    else:
        print(count,"Printing inside while loop")
    count += 1
print("Code Ended!")
print(count)
```

### Use of **continue**

```python
count = 0
while count < 10:
    if count == 5:
        count += 1
        continue
        print("Print after continue")
    else:
        print(count,"printing in while -> else")
    count += 1
print("Code Ended!")
print(count)
```

### Infinity loop due to **inc/dec issue**

```python
count = 9 # Value Initialization
while count >= 0: # Condition
    # Block of Code
    print(count,"Printing inside while block")
    count += 1 # Decrement
print("Code Ended!") # code line after while block
```

### Infinity loop using **True**

```python
total = 0
while True:
    num = int(input())
    if num < 0:
        break
    else:
        total += num
print(total)
```

```python
#Keep taking integers input from user until he enters "STOP".
#Print the number of odd and even numbers entered by the user.
even_count = 0
odd_count = 0
while True:
    number = input("Please enter a number: ")
    if number == "STOP":
        break
    elif number.isnumeric():
        if int(number)%2==0:
            even_count += 1
        else:
            odd_count += 1
    else:
      print('Invalid string')
print("Even Count: ",even_count)
print("Odd Count: ",odd_count)
```

### range() function


*   Generates numbers that is specified inside the parenthesis
*   The start is inclusive and end is exclusive
*   **format**=> **range(start, stop, step)**
*   step denotes by what value in each iteration the iterator will increase/decrease
*   Default value for start = 0
*   Deafault value for step = *1*

```python
# Write a porgram that checks if an integer is an prime number or not
number = int(input()) # 5

if number == 0 or number == 1:
  print("Not prime")
else:
  divisor_counter = 0
  for i in range(1, number + 1): # 1 2 3 4 5
    if number % i == 0:
      divisor_counter += 1
  if divisor_counter == 2:
    print("prime",divisor_counter)
  else:
    print("Not prime", divisor_counter)
```
