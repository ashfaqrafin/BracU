```python
# Task -2

class Student:
    def __init__(self,name,id,cgpa):
        self.name = name
        self.id = id
        self.cgpa = cgpa
    def setId(self,id):
        self.id = id


class Department:
    def __init__(self,name):
        self.name = name
        self.students = []
    def addStudent(self,*student):
        for i in range(len(student)):
            for k in range(len(self.students)):
                if self.students[k].id == student[i].id:
                    print(f"Student with ID {student[i].id} already exists,  Please try with another ID")
                    return
            print(f"Welcome to {self.name} department, {student[i].name}")
            self.students.append(student[i])
    def findStudent(self,id):
        for i in range(len(self.students)):
            if self.students[i].id == id:
                print("Student info:")
                print("Name:",self.students[i].name)
                print("ID:",self.students[i].id)
                print("CGPA:",self.students[i].cgpa)
                return
        print("Student with this ID doesn't exist, Please give a valid ID")
    def details(self):
        print(f"Department Name: {self.name}")
        print("Number of student:",len(self.students))
        print("Details of the students:")
        for i in range(len(self.students)):
            print(f"Name: {self.students[i].name}, ID: {self.students[i].id}, CGPA: {self.students[i].cgpa}")


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
# Task-3
class Vaccine:
    def __init__(self, name, country, days_between_doses):
        self.name = name
        self.country = country
        self.days_between_doses = days_between_doses

class Person:
    def __init__(self, name, age, occupation="General Citizen"):
        self.name = name
        self.age = age
        self.occupation = occupation
        self.vaccine = None
        self.doses_taken = 0

    def pushVaccine(self, vaccine, dose="1st Dose"):
        if self.age < 25 and self.occupation == "Actor":
            print(f"Sorry {self.name}, Minimum age for taking vaccines is 25 years now.")
            return
        if self.vaccine is not None and self.vaccine.name != vaccine.name:
            print(f"Sorry {self.name}, you can’t take 2 different vaccines")
            return
        self.vaccine = vaccine
        self.doses_taken += 1
        print(f"{dose} dose done for {self.name}")

    def showDetail(self):
        print(f"Name: {self.name} Age: {self.age} Type: {self.occupation}")
        if self.vaccine is not None:
            print(f"Vaccine name: {self.vaccine.name}")
            if self.doses_taken == 1:
                print("1st dose: Given\n2nd dose: Please come after {} days".format(self.vaccine.days_between_doses))
            else:
                print("1st dose: Given\n2nd dose: Given")

astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student") 
print("=================================") 
p1.pushVaccine(astra) 
print("=================================") 
p1.showDetail() 
print("=================================") 
p1.pushVaccine(sin, "2nd Dose") 
print("=================================") 
p1.pushVaccine(astra, "2nd Dose") 
print("=================================") 
p1.showDetail() 
print("=================================") 
p2 = Person("Carol", 23, "Actor") 
print("=================================") 
p2.pushVaccine(sin) 
print("=================================") 
p3 = Person("David", 34) 
print("=================================") 
p3.pushVaccine(modr) 
print("=================================") 
p3.showDetail() 
print("=================================") 
p3.pushVaccine(modr, "2nd Dose")
```
