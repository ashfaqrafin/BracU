```python
# Task-1
class NikeBangladesh:
    branches=[]
    total_stock={"Air Jordan":0,"Cortez":0,"Zoom Kobe":0} 
    total_sold=0
    def __init__(self, location):
        self.location = location
        self.products = {"Air Jordan": 0, "Cortez": 0, "Zoom Kobe": 0}
        NikeBangladesh.branches.append(location)
    def details(self):
        print(f"Nike {self.location} outlet:")
        print("Products Currently Stocked:")
        print(self.products)
    def restockProducts(self, products):
        for key in products:
            self.products[key] += products[key]
            NikeBangladesh.total_stock[key] += products[key]
    def productSold(self, products):
        for key in products:
            self.products[key] -= products[key]
            NikeBangladesh.total_stock[key] -= products[key]
            NikeBangladesh.total_sold += products[key]
    @classmethod
    def status(cls):
        print("Nike Bangladesh Status:")
        print(f"Branches Opened: {cls.branches}")
        print(f"Cureently Stocked Products\n{cls.total_stock}")
        print(f"Sold: {cls.total_sold}")



print("xxxxxxxxxxxxxx1xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
dhaka = NikeBangladesh("Dhaka Banani")
chittagong = NikeBangladesh("Chittagong GEC")
print("xxxxxxxxxxxxxx2xxxxxxxxxxxxxxxx")
dhaka.details()
print("xxxxxxxxxxxxxx3xxxxxxxxxxxxxxxx")
chittagong.details()
print("xxxxxxxxxxxxxx4xxxxxxxxxxxxxxxx")
dhaka.restockProducts(
{"Air Jordan":1200,"Cortez":200,"Zoom Kobe":200})
chittagong.restockProducts(
{"Air Jordan":1000,"Cortez":250,"Zoom Kobe":100})
print("xxxxxxxxxxxxxx5xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
print("xxxxxxxxxxxxxx6xxxxxxxxxxxxxxxx")
dhaka.productSold({"Air Jordan":760,"Cortez":90})
chittagong.productSold({"Air Jordan":520,"Zoom Kobe":70})
print("xxxxxxxxxxxxxx7xxxxxxxxxxxxxxxx")
NikeBangladesh.status()
```

```python
# Task-3

class SultansDine:
    branches=[] 
    total_sell=0 
    def __init__(self, branch):
        self.branch = branch
        self.sell = 0
        SultansDine.branches.append(self)
    def sellQuantity(self, quantity):
        if quantity < 10:
            self.sell += quantity*300
            SultansDine.total_sell += quantity*300
        elif quantity < 20:
            self.sell += quantity*350
            SultansDine.total_sell += quantity*350
        else:
            self.sell += quantity*400
            SultansDine.total_sell += quantity*400
    def branchInformation(self):
        print(f"Branch Name: {self.branch}")
        print(f"Branch Sell: {self.sell} Taka")
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {len(cls.branches)}")
        print(f"Total Sell: {cls.total_sell} Taka")
        for i in (cls.branches):
            print(f"Branch Name: {i.branch}, Branch Sell: {i.sell} Taka, Branch consists of total sell's: {i.sell/cls.total_sell*100:.2f}%")

SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```

```python
from math import exp, factorial

# Poisson probability calculation
def poisson_probability(lam, k):
    return (exp(-lam) * lam**k) / factorial(k)

# Calculate the probability of 5 or fewer defects
probability_5_or_fewer = sum(poisson_probability(5, k) for k in range(6))

# Calculate the probability of more than 5 defects
probability_more_than_5 = 1 - probability_5_or_fewer

probability_more_than_5
```
