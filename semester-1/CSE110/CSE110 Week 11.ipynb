### **Read Mode:**
```
 Can only read lines of the file. Cannot write or change content of the file.
```
**This is the default mode that a file is opened in if file mode not specified.**

This mode only works if the file exists.


````python
# f = open("/content/file1.txt", "r")
# print(f)
# file_read = f.read()
# print(file_read)
# f.close()

f2 = open("file1.txt")
print(f2)
f2_read = f2.read()
print(f2_read)
f2.close()
````

### **Write Mode:**

```
Can only write lines to the file.
```
If the file did not exist previously then a new file is created when this mode is specified.

If file exists before then the content of the file is deleted and writing starts form the begining of the file.

**You cannot read the file in this mode.**


````python
file2 = open("/content/write1.txt", "w")
print(file2)
file2.write("Okay start writing this")
file2.close()

file2 = open("write1.txt",'r')
read = file2.read()
print(read)
file2.close()

file1 = open("/content/file1.txt",'w')
file1.write("wrting this \nNow")
file1.close()
file1 = open("file1.txt",'r')
read = file1.read()
print(read)
file1.close()
````


### **Append Mode:**
```
Can add content to a file after the previously written content.
```
If the file did not exist previously then a new file is created when this mode is specified.

**You cannot read the file in this mode.**


````python
file1 = open("/content/file1.txt",'a')
file1.write("\nWriting again at the end in file 1 using append")
file1.flush()
file1.close()

file1 = open("file1.txt",'r')
read = file1.read()
print(read)
file1.close()
````

#### **Other modes**


*   Create mode/ **x mode**: Creates a new file only if a file in the same name did not exist before.
*   Text mode/ **t mode**: We can add the t mode to other modes. This is used to read/write/append the file as text. This is the default mode, all files file operations are done in text mode.
*   Binary mode/ **b mode**: This is used to read binary files e.g: Images(jpg, jpeg, png), PDFs etc.



````python
file1 = open("file1.txt",'at')
file1.write("\nWriting at the end in file 1")
file1.close()
file1 = open("file1.txt",'rt')
read = file1.read()
print(read)
file1.close()
````

**Closing a file is important**

1.     If you do not close a file after you are done reading/writing, the operating system level resource for file management remain occupied. There is a limit in how many such resource you can occupy in a program. So if you do not close the files and keep reading new files, then your program will crash or will be killed by the operating system.
2.   Finally, if you do not close the file after you are done writing, then your changes may not be persisted in the disk as the operating system often copy data back from memory buffer to disk only during file close operation. **There is another method call flush() though that you can use to force update of the disk version of the file before file close.**
3. Another alternative to closing your files can be using **with open()**

````python
with open("file1.txt",'a') as f:
  f.write("\nExecuting 'with'")

with open("file1.txt",'r') as f:
  print(f.read())
#f.write("Trying to write")
````

### Some other methods of file IO
1. readline() : Reads each line one by one as a string.
2. readlines() : Reads the whole .txt file as a list. The next lines are read as \n
3. writelines() : Will write lines from and iterable object to the text file. You need to add the next line or \n to the strings yourself.

````python
f = open("file1.txt", "r")
line1  = f.readlines()
print(line1)
print(type(line1))
for i in range(len(line1)):
  print(line1[i], end = "")
f.close()
````

````python
f = open("write1.txt", "a")
f.writelines(['\nwriting line 4\n', 'line 2\n', 'ending with this'])
f.close()
f = open("write1.txt", "r")
print(f.read())
f.close()
````

It is however risky to use readLines, and read without any character count because if your program is processing a large file then the content of the whole file will be loaded into the memory and your program will consume too much memory as a result. Then in the worst case when the file is too large, the program will crash.

The best strategy of file reading is to read only as much data as you need at a time and then let go of that data when you do not need it any more.

````python
f = open("write1.txt", "r")
char = f.read(1)
# flag = False
# newStr = ""
while char != "":
  print(char, end="")
  char = f.read(1)
f.close()
````

````python
f = open("write1.txt", "r")
# print(f.readline())
# print(f.readline())
line = f.readline()
while line != "":
  print(line, end = "")
  line = f.readline()
f.close()
````

#### **seek(), tell() & truncate()**

These functions are used to in reference to the position of things inside the file.

1. seek() : allows you to move to a certain position specified inside the seeks argument.

2. tell() : This function return the current character being read in the file. Using tell() you can learn which character you are on using the seek() function.

3. truncate() : Limits the number of characters of your file. To use this function you MUST open file in w/a mode.

Refer to this: [using seek() to go forward and backward](https://www.geeksforgeeks.org/python-seek-function/)

````python
# seek(offset, from_what)
# from_what 0 -> beginning, 1 - > current position, 2 - > end of file
f = open("file3.txt", "w")
f.write("ABCDEFGH")
f.close()

f = open("file3.txt", "r")
f.seek(5)
print("After seeking",f.tell())

reading = f.read(5)
print("Reading the next 5 characters",reading)

print("After reading",f.tell())
f.close()

print("_______________________")

f =open('file3.txt','r')
print(f.read())
f.close()

f = open("file3.txt", "a")
f.truncate(6)
f.close()

f = open("file3.txt", "r")
line1 = f.read()
print("Truncated",line1)
f.close()
````
