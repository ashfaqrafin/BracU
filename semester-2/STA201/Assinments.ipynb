```python
# Assingment-1
import pandas as pd
import matplotlib.pyplot as plt

# Create a vector of transmission values
transmission = ["Automatic", "Automatic", "Automatic", "Manual", "Automatic", "Manual", "Automatic", "Manual"]

# Create a frequency table using the pandas value_counts() function
freq_table = pd.Series(transmission).value_counts()

# Create a bar chart using the pandas plot() function
freq_table.plot(kind='bar', title='Bar Chart of Transmission', xlabel='Transmission', ylabel='Frequency')
plt.show()

# Create a pie chart using the pandas plot() function
(freq_table / freq_table.sum() * 100).plot(kind='pie', autopct='%1.1f%%', title='Pie Chart of Transmission')
plt.show()
```
