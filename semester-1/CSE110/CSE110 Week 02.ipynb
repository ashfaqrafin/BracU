```python
number1 = int(input("Enter a number: "))
number2 = int(input("Enter a number: "))
if number1 >= number2:
  print("This is inside if block")
  if number1 > number2:
    print("This is inside if - nested if block")
    print(number1, "is greater than", number2)
  else:
    print(number1, "is equal to",number2)
    print("This is inside if - nested else block")
else:
  print("This is inside else block")
  print(number1, "is less than", number2)
print("End of program")
```

```python
number1 = int(input("Enter a number: "))
number2 = int(input("Enter a number: "))
if number1 > number2:
  print("This is inside if block")
  print(number1, "is greater than", number2)
else:
  print("This is inside else block")
  if number1 < number2:
      print("This is inside else - nested if block")
      print(number1, "is less than", number2)
  else:
      print(number1, "is equal to",number2)
      print("This is inside else - nested else block")

print("End of program")
```
