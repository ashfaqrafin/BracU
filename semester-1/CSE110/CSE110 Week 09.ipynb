#### Array Operations

```python
arr = np.array([10, 20, 30, 40, 50])

# read/access an element
val = arr[3]
print(val)

# write/modify/update an element
arr[2] = 100
print(arr)
```

```python
arr = np.array([10, 12, 15, 18, 21])

# traversal / iteration
print("Iteration Using While Loop:")
idx = 0
while idx < arr.size: # idx < len(arr)
    print(arr[idx])
    idx += 1

print()

print("Iteration Using For Loop:")
for idx in range(arr.size): # range(len(arr))
    print(arr[idx])
```

#### Resize an Array
We know that unlike lists, we can not append element in arrays.

But if you want to add more elements, you need to resize the array.

To resize an array,

1.   Create a new array with increased length
2.   Copy the elements from the old array to new array
3.   Then, insert your new element values



```python
arr1 = np.array([10, 15, 20, 45, 50])
print(arr1.size)

# let's say, you want to add another element 60 in the array
newSize = arr1.size + 1
arr2 = np.zeros(newSize, dtype = "int")
for i in range(arr1.size):
    arr2[i] = arr1[i]
# print(arr2)
arr2[newSize-1] = 60
print(arr2)
```

### Array Searching: Linear Search

Linear Search/Sequential Search is one of the popular searching techniques in a sequence.

It searches for a particular element sequentially and returns its index.

```python
def linearSearch(arr, val):
    for idx in range(arr.size):
        if arr[idx] == val:
            return "Value found at index " + str(idx)
    return "Value not found"

myArr = np.array([10, 50, 27, -5, 3, 12, -2, 31])
print(linearSearch(myArr, 12))
print(linearSearch(myArr, 25))
```

### Array Sorting: Bubble Sort
Sorting means rearranging values in a particular order.
<br>

Two types:
1.   Ascending Order / Increasing Order / Smallest to Largest / "Sort" (by default)
2.   Descending Order / Decreasing Order / Non-increasing Order / Largest to Smallest

In this class, we will learn Bubble Sorting Technique.
*   main idea is to check and **swap** between consecutive two index values
*   starting from the right-most index, the values from each index get sorted one by one after each iteration

Before moving on to the code, let's watch some visualization from [here](https://visualgo.net/en/sorting).


```python
def bubbleSortAscending(arr):
    for i in range(arr.size-1):
        for j in range(arr.size-1-i):
            if arr[j+1] < arr[j]:
                # swapping
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
    return arr

myArr = np.array([10, 50, 27, -5, 3, 12, -2, 31])
print("Before Sorting:", myArr)

myArr = bubbleSortAscending(myArr)
print("After Sorting: ", myArr)
```

```python
def bubbleSortDescending(arr):
    for i in range(arr.size-1):
        for j in range(arr.size-1-i):
            if arr[j+1] > arr[j]:
                # swapping
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
    return arr

myArr = np.array([10, 50, 27, -5, 3, 12, -2, 31])
print("Before Sorting:", myArr)

myArr = bubbleSortDescending(myArr)
print("After Sorting: ", myArr)
```
