#### Resize an Array
We know that unlike lists, we can not append element in arrays.

But if you want to add more elements, you need to resize the array.

To resize an array,

1.   Create a new array with increased length
2.   Copy the elements from the old array to new array
3.   Then, insert your new element values



```python
arr1 = np.array([10, 15, 20, 45, 50])
print(arr1.size)
print(arr1)
# let's say, you want to add another element 60 in the array
newSize = arr1.size + 1
arr2 = np.zeros(newSize, dtype = "int")
print(arr2, "initial")
for i in range(arr1.size): #0, 1, 2, 3, 4
    arr2[i] = arr1[i]
print(arr2)
arr2[newSize-1] = 60
print(arr2)
```

### Array Searching: Linear Search

Linear Search/Sequential Search is one of the popular searching techniques in a sequence.

It searches for a particular element sequentially and returns its index.

```python
def linearSearch(arr, val):
    for idx in range(arr.size):
        if arr[idx] == val:
            return "Value found at index " + str(idx)
    return "Value not found"

myArr = np.array([10, 50, 27, -5, 3, 12, -2, 31])
print(linearSearch(myArr, 12))
print(linearSearch(myArr, 25))
```

### Array Sorting: Bubble Sort
Sorting means rearranging values in a particular order.
<br>

Two types:
1.   Ascending Order / Increasing Order / Smallest to Largest / "Sort" (by default)
2.   Descending Order / Decreasing Order / Non-increasing Order / Largest to Smallest

In this class, we will learn Bubble Sorting Technique.
*   main idea is to check and **swap** between consecutive two index values
*   starting from the right-most index, the values from each index get sorted one by one after each iteration

image.png

Before moving on to the code, let's watch some visualization from [here](https://visualgo.net/en/sorting).


```python
def bubbleSortAscending(arr):
    for i in range(arr.size): #range(4)
        for j in range(arr.size-1-i): #range(4), range(3), range(2)
            if arr[j] > arr[j+1]:
                # swapping
                temp = arr[j]
                arr[j] = arr[j+1]
                arr[j+1] = temp
    return arr

myArr = np.array([-2, 45, 0, 11, -9])
print("Before Sorting:", myArr)

myArr = bubbleSortAscending(myArr)
print("After Sorting: ", myArr)
```

#### Matrix Addition
Screenshot 2023-11-28 013235.png

```python
A = np.array([[1, 3], [1, 0], [1, 2]])
B = np.array([[0, 0], [7, 5], [2, 1]])
print(A)
print()
print(B)
print()

a_row = A.shape[0]
a_col = A.shape[1]
b_row = B.shape[0]
b_col = B.shape[1]

# matrix shapes must be the same
if (a_row != b_row or a_col != b_col):
    print("Addition Not Possible.")

else: # same shape
    C = np.zeros(shape = (a_row, a_col), dtype = "int")
    for row in range(a_row):
        for col in range(a_col):
            C[row][col] = A[row][col] + B[row][col]
    print(C)
```

#### Matrix Row Switch (Using Loop)

```python
A = np.array([[1, -1, 2, 0], [-2, 3, 7, 1], [4, 5, -6, -3]])

# switch row 1 and 2 [row starts from 0]
firstRow = 1
secondRow = 2

col = A.shape[1]
for i in range(col):
    temp = A[firstRow][i]
    A[firstRow][i] = A[secondRow][i]
    A[secondRow][i] = temp
print(A)
```

#### Matrix & Matrix Multiplication (Using Loop)



*   column number of first matrix = row number of second matrix
*   resulting matrix  = row number of first matrix x column number of second matrix

```python
A=np.array( [[0,5],[1,2],[3,4]] )
B=np.array( [[3,2,1],[1,2,4]] )
print("A(3,2):")
print(A)
print()
print("B(2,3):")
print(B)
print()

Ar_len = A.shape[0]
Ac_len = A.shape[1]
Br_len = B.shape[0]
Bc_len = B.shape[1]

if Ac_len != Br_len:
    print("Multiplication Not Possible.")
else:
    C = np.zeros( shape=(Ar_len,Bc_len), dtype=int)
    Cr_len = C.shape[0]
    Cc_len = C.shape[1]
    for i in range(Cr_len): #3
        for j in range(Cc_len):  #3
            total = 0
            for k in range(Br_len): # range(Ac_len)
                total += A[i][k] * B[k][j]
            C[i][j] = total
    print("A X B = C(3,3):")
    print(C)
```
