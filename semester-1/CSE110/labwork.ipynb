```python
given=input()
flag=True
for i in given:
    if(not (ord(i)==48 or ord(i)==49)):
        flag=False
        break

print({True:"Binary number",False:"Not a binary number"}[flag])
```

```python
given=input()
vowels=['a','e','i','o','u']
count_vowels=0
count_consonants=0
for i in given:
    if(not 65<=ord(i)<=90 and not 97<=ord(i)<=122):
        continue
    if(i in vowels):
        count_vowels+=1
    else:
        count_consonants+=1

if((not count_vowels%3 and count_vowels) and (not count_consonants%5 and count_consonants)):
    print("Aaarr! Me Plunder!!")
else:
    print("Blimey! No Plunder!!")

# print({True:"Aaarr! Me Plunder!!",False:"Blimey! No Plunder!!"}[bool((not count_vowels%3 and count_vowels) and (not count_consonants%5 and count_consonants))])
```
