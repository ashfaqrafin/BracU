```python
given=input()
ans=""
count=0
for i in given:
    if(not 65<=ord(i)<=90 and not 97<=ord(i)<=122):
        ans+=i
        continue
    if(not count&1):
        ans+={True:chr(ord(i)-32),False:i}[97<=ord(i)<=122]
    else:
        ans+={True:chr(ord(i)+32),False:i}[65<=ord(i)<=90]
    count+=1
        
print(ans)
```

```python
given=input()
i=0
while(given[i]!=" "):
    i+=1
first,second=0,i+1
ans=""
while(first!=i-1 and second!=len(given)):
    ans+=(given[first]+given[second])
    first+=1
    second+=1
ans+={True:given[slice(first,i-1)],False:given[slice(second,len(given))]}[first<i-1]
print(ans)
```
