```python
# Task-1
class Employee:
    def __init__(self,name,id,position,etype="regular",salary=30000):
        self.name=name
        self.id=id
        self.position=position
        self.type=etype
        self.salary=salary
    def employeeDetails(self):
        print(f"Name: {self.name}, Dept {self.position}")
        print(f"Employee id: {self.id}, Salary: {self.salary}")
    def workDistribution(self,position):
        if(position=="Human Resource"):
            print("Collect work distribution details from the manager.")
        else:
            print("Collect work distribution loads from the HR department.")
    def increment(self):
        if(self.type=="regular"):
            self.salary+=(self.salary*0.1)
        else:
            self.salary+=(self.salary*0.15)
class Foreign_employee(Employee):
    def __init__(self,name,id,position):
        super().__init__(name,id,position,etype="foreign")
    def employeeDetails(self):
        super().employeeDetails()
        print("Employee Type: Foreign")
    
class Part_time_employee(Employee):
    def __init__(self,name,id,position):
        super().__init__(name,id,position,etype="part-time",salary=15000)
    def employeeDetails(self):
        super().employeeDetails()
        print("Employee Type: Part Time")
    def increment(self): 
        print("Sadly, there is no increment for the part time employees!!")

print("1------------------------------------------") 
emp1=Employee("Nawaz Ali", 102, "Marketing") 
print("2------------------------------------------") 
emp1.employeeDetails() 
print("3------------------------------------------") 
emp1.workDistribution("Marketing") 
print("4------------------------------------------") 
emp1.increment()
emp1.employeeDetails() 
print("5------------------------------------------") 
f_emp=Foreign_employee("Nadvi", 311, "Human Resource") 
f_emp.employeeDetails() 
print("6------------------------------------------") 
f_emp.workDistribution("Human Resource") 
print("7------------------------------------------")
f_emp.increment()
f_emp.employeeDetails() 
print("8------------------------------------------") 
p1_emp=Part_time_employee("Asif", 210, "Sales") 
p2_emp=Part_time_employee("Olive", 223, "Accounts") 
print("9------------------------------------------") 
p1_emp.employeeDetails() 
print("10------------------------------------------") 
p1_emp.workDistribution("Sales") 
print("11------------------------------------------") 
p2_emp.increment()
print("12------------------------------------------") 
p2_emp.employeeDetails()
```

```python
# Task-2
class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60
    def examSyllabus(self):
        return "Maths , English"
    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self,marks,time,*sub):
        super().__init__(marks)
        self.time=time
        self.subs=sub
    def __str__(self):
        return f"Marks: {self.marks} Time: {self.time} minutes Number of Parts: {len(self.subs)+2}"
    def examSyllabus(self):
        return super().examSyllabus()+",".join(self.subs)
    def examParts(self):
        return super().examParts()+"\n".join(f"Part {i+2} - {self.subs[i]}" for i in range(len(self.subs)))
    
        
    
engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture =ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
# Task-3
class PokemonBasic:
    def __init__(self, name = 'Default', hp = 0,weakness = 'None', type = 'Unknown'):
        self.name = name
        self.hit_point = hp
        self.weakness = weakness
        self.type = type
    def get_type(self):
        return 'Main type: ' + self.type
    def get_move(self):
        return 'Basic move: ' + 'Quick Attack'
    def __str__(self):
        return "Name: " + self.name + ", HP: " +str(self.hit_point) + ", Weakness: " + self.weakness
class PokemonExtra(PokemonBasic):
    def __init__(self,name,hp,weakness,type1,type2=None,moves=None):
        super().__init__(name,hp,weakness,type1)
        self.type2=type2
        self.moves=moves
    def get_type(self):
        forReturn=super().get_type()
        if(self.type2):
            return forReturn+f", Secondary type: {self.type2}"
        else:
            return forReturn
    def get_move(self):
        forReturn=super().get_move()
        if(self.moves):
            moves=", ".join(self.moves)
            return forReturn+f"\nOther move: {moves}"
        else:
            return forReturn
        

print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())
print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water',
'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water',
'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
# Task-4

class Cakes:
    order_list={}
    feedbacks={}
    def __init__(self,flavor,weight,cost=1200,sweetness="Moderate sugar",type="Whipped Cream"):
        self.flavor=flavor+" Cake"
        self.weight=weight
        self.price=cost*weight/1000
        self.sweetness=sweetness
        self.type=type
        if self.flavor+f" {weight}gm" in Cakes.order_list:
            Cakes.order_list[self.flavor+f" {weight}gm"]+=1
        else:
            Cakes.order_list[self.flavor+f" {weight}gm"]=1
    def cake_details(self):
        print(f"Flavor: {self.flavor}, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness}, Cream Type: {self.type}")
        print(f"Price: {self.price} Taka")
    def add_customization(self,sweetness,type):
        self.sweetness=sweetness+" sugar"
        self.type=type
    @classmethod
    def give_feedbacks(cls,flavor,compliment):
        if(flavor in cls.feedbacks):
            cls.feedbacks[flavor].append(compliment)
        else:
            cls.feedbacks[flavor]=[compliment]
        print("Thanks for your valuable feedback!")
    @classmethod
    def show_feedbacks(cls):
        print(cls.feedbacks)
class Cheese_Cakes(Cakes):
    def __init__(self,flavor,weight,isit="baked"):
        super().__init__(flavor+" Cheese",weight,1500)
        self.type="Cream Cheese"
        self.isit=isit
        if(self.flavor in Cakes.feedbacks):
            self.price-=self.price*0.1
    def cake_details(self):
        print(f"Flavor: {self.flavor}, Weight: {self.weight} gm")
        print(f"Sweetness: {self.sweetness}, Cream Type: {self.type}")
        print(f"Cake Type:{self.isit},Price: {self.price} Taka")
    def add_customization(self):
        print("Sorry! No customization available for cheese cakes.")
    @classmethod
    def give_feedbacks(cls, flavor, compliment):
        super().give_feedbacks(flavor, compliment)
        print(f"You will get 10% discount for your next purchase!")

    


order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
