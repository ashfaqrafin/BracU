Assume, you have been given a list and a number (N). Now, create a list by using the input list where range goes from 1 to N.

**Given Input:** \\
`list1 = ['p','q','r','s']` \\
`N = 3` \\

**Sample Output:** \\
`['p1', 'q1', 'r1', 's1', 'p2', 'q2', 'r2', 's2', 'p3', 'q3', 'r3', 's3']`

```python
list1 = ['p','q','r','s']
N = int(input())
new_list = []
for i in range(1, N+1):
  for j in range(len(list1)):
    temp = list1[j] + str(i)
    new_list.append(temp)

print(new_list)
```

```python
# function to print everything of a list in a sequence
list = [10, 20, [5, 6, [9, 10, 11, [6, [7, 6], 8]], 6, "Yes"], "Hello World"]

def recursiveListPrint(argList, tabCount):
  itemCount = len(argList)
  for i in range(0, itemCount, 1):
    item = argList[i]
    if type(item) == type([]):
      nestedList = item
      recursiveListPrint(nestedList, tabCount + 1)
    else:
      for i in range(0, tabCount, 1):
        print('\t', end="")
      print(item)

recursiveListPrint(list, 0)
```

Problem 3: suppose we have a list as input that may have other lists inside it. What we want is to copy the elements from inside list to the outside list directly.

Example:

Input: [2, [3, 4], 5, ["Merry", "John"]]

Output: [2, 3, 4, 5, "Merry", "John"]

```python
given = [2, [3, 4, 6], 5, ["Merry", "John"], "Peter"]
output_list = []
for i in range(len(given)):
  if type(given[i]) == list:
    for j in range(len(given[i])):
      output_list.append(given[i][j])
  else:
    output_list.append(given[i])
print(output_list)
```
